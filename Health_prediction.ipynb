```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

from xgboost import XGBClassifier
from catboost import CatBoostClassifier

from imblearn.over_sampling import SMOTE


# -----------------------------
# TARGETS
# -----------------------------
binary_targets = [
    'has_diabetes',
    'has_hypertension',
    'has_heart_disease',
    'has_kidney_disease',
    'has_liver_disease',
    'has_asthma'
]

# -----------------------------
# MODEL STORAGE
# -----------------------------
logistic_models = {}
dt_models = {}
nb_models = {}
xgb_models = {}
cat_models = {}

results = []


# -----------------------------
# TRAIN MODELS FOR EACH TARGET
# -----------------------------
for target in binary_targets:
    print(f"\n================ MODELING: {target.upper()} ================\n")

    y = df[target]
    X = df.drop(columns=binary_targets + ['patient_id'], errors='ignore')

    # Train-Test Split (Stratified)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.20, random_state=42, stratify=y
    )

    # Scaling (needed for LR, DT, NB)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)


    # =====================================================
    # 1ï¸âƒ£ LOGISTIC REGRESSION (BASELINE)
    # =====================================================
    log_model = LogisticRegression(
        max_iter=2000,
        class_weight="balanced"
    )
    log_model.fit(X_train_scaled, y_train)

    log_probs = log_model.predict_proba(X_test_scaled)[:, 1]
    log_preds = (log_probs > 0.5).astype(int)

    log_acc = accuracy_score(y_test, log_preds)
    log_f1 = f1_score(y_test, log_preds)
    log_auc = roc_auc_score(y_test, log_probs)

    print("ðŸ“Œ Logistic Regression")
    print(classification_report(y_test, log_preds))

    logistic_models[target] = log_model


    # =====================================================
    # 2ï¸âƒ£ DECISION TREE
    # =====================================================
    dt = DecisionTreeClassifier(
        max_depth=6,
        class_weight="balanced",
        random_state=42
    )
    dt.fit(X_train_scaled, y_train)

    dt_probs = dt.predict_proba(X_test_scaled)[:, 1]
    dt_preds = (dt_probs > 0.5).astype(int)

    dt_acc = accuracy_score(y_test, dt_preds)
    dt_f1 = f1_score(y_test, dt_preds)
    dt_auc = roc_auc_score(y_test, dt_probs)

    print("\nðŸ“Œ Decision Tree")
    print(classification_report(y_test, dt_preds))

    dt_models[target] = dt


    # =====================================================
    # 3ï¸âƒ£ NAIVE BAYES
    # =====================================================
    nb = GaussianNB()
    nb.fit(X_train_scaled, y_train)

    nb_probs = nb.predict_proba(X_test_scaled)[:, 1]
    nb_preds = (nb_probs > 0.5).astype(int)

    nb_acc = accuracy_score(y_test, nb_preds)
    nb_f1 = f1_score(y_test, nb_preds)
    nb_auc = roc_auc_score(y_test, nb_probs)

    print("\nðŸ“Œ Naive Bayes")
    print(classification_report(y_test, nb_preds))

    nb_models[target] = nb


    # =====================================================
    # 4ï¸âƒ£ XGBOOST + SMOTE
    # =====================================================
    sm = SMOTE(random_state=42)
    X_train_res, y_train_res = sm.fit_resample(X_train_scaled, y_train)

    scale_pos_weight = (len(y_train_res) - sum(y_train_res)) / sum(y_train_res)

    xgb = XGBClassifier(
        n_estimators=350,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric='logloss',
        scale_pos_weight=scale_pos_weight,
        random_state=42
    )

    xgb.fit(X_train_res, y_train_res)

    xgb_probs = xgb.predict_proba(X_test_scaled)[:, 1]
    xgb_preds = (xgb_probs > 0.30).astype(int)   # tuned threshold

    xgb_acc = accuracy_score(y_test, xgb_preds)
    xgb_f1 = f1_score(y_test, xgb_preds)
    xgb_auc = roc_auc_score(y_test, xgb_probs)

    print("\nðŸ“Œ XGBoost")
    print(classification_report(y_test, xgb_preds))

    xgb_models[target] = xgb


    # =====================================================
    # 5ï¸âƒ£ CATBOOST
    # =====================================================
    cat = CatBoostClassifier(
        iterations=300,
        depth=6,
        learning_rate=0.05,
        loss_function='Logloss',
        eval_metric='AUC',
        verbose=0,
        random_state=42
    )

    # CatBoost prefers unscaled data
    cat.fit(X_train, y_train)

    cat_probs = cat.predict_proba(X_test)[:, 1]
    cat_preds = (cat_probs > 0.30).astype(int)

    cat_acc = accuracy_score(y_test, cat_preds)
    cat_f1 = f1_score(y_test, cat_preds)
    cat_auc = roc_auc_score(y_test, cat_probs)

    print("\nðŸ“Œ CatBoost")
    print(classification_report(y_test, cat_preds))

    cat_models[target] = cat


    # =====================================================
    # SAVE METRICS
    # =====================================================
    results.append({
        "target": target,

        "logistic_acc": log_acc, "logistic_f1": log_f1, "logistic_auc": log_auc,
        "dt_acc": dt_acc, "dt_f1": dt_f1, "dt_auc": dt_auc,
        "nb_acc": nb_acc, "nb_f1": nb_f1, "nb_auc": nb_auc,

        "xgb_acc": xgb_acc, "xgb_f1": xgb_f1, "xgb_auc": xgb_auc,
        "cat_acc": cat_acc, "cat_f1": cat_f1, "cat_auc": cat_auc
    })


# -----------------------------
# FINAL COMPARISON TABLE
# -----------------------------
comparison = pd.DataFrame(results)

print("\n================ FINAL MODEL COMPARISON ================\n")
print(comparison)
```
